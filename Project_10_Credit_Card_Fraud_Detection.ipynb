**Importing the libraries of the dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset to the pandas dataframe
credit_card_dataset = pd.read_csv('/content/creditcard.csv')
```

```python
# Loading the first 5 rows of the dataframe
credit_card_dataset.head()
```

```python
# Loading the last 5 rows of the dataframe
credit_card_dataset.tail()
```

```python
# Loading some information about the dataset
credit_card_dataset.describe
```

```python
# Checking the number of the null values in the dataframe
credit_card_dataset.isnull().sum()
```

```python
# Distribution of the legit transactions and fraudlent transactions
credit_card_dataset['Class'].value_counts()
```

**The loaded data is highly unbalanced.**

**Here, 0 --> Normal Transactions
      1 --> Fraudlent Transactions**
      

```python
# separating the data for analysis
legit = credit_card_dataset[credit_card_dataset.Class == 0.0]
fraud = credit_card_dataset[credit_card_dataset.Class == 1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the legit data
legit.Amount.describe()
```

```python
# Statistical measures of the fraud data
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_dataset.groupby('Class').mean()
```

**Since the dataframe is unbalanced so we are doing Under Sampling of the data**

**Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions**



```python
legit_sample = legit.sample(n=201)
```

**Concatenating two DataFrames**

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
# Now the data is uniformly distributed and being balanced data
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**Splitting the data into Features & Targets**

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
print(Y)
```

**Split the data into Training data & Testing Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**

**Logistic Regression**

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

**Model Evaluation**

# **Accuracy** **Score**

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
